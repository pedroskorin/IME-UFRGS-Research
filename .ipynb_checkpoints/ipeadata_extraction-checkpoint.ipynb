# Extracting data from IPEA Data
### Using ipeadatapy

```python
#pip install ipeadatapy
import ipeadatapy as idpy
import pandas as pd
```

```python
# Apresentação de todas as variáveis Macroeconômicas disponíveis no Ipea Data
idpy.themes(macro=1)
```

```python
# As características desejadas das séries temporais podem ser alteradas aqui
test = idpy.metadata(big_theme = 'Macroeconômico', country = "BRA", frequency = "Mensal")
test.head()
```

```python
# Filtragem de séries que foram atualizadas pela última vez pelo menos em dezembro de 2020 (séries recentes)

filt_lu = [0]*len(test)

# Last Update
for i in range(len(test)):
    ti = datetime.strptime(test["LAST UPDATE"][i][0:10], "%Y-%m-%d")
    if ti >= datetime.strptime("2020-12-01", "%Y-%m-%d"):
        filt_lu[i] = True
    else:
        filt_lu[i] = False

test_lu = test.iloc[filt_lu]
```

```python
# Parte demorada do código, captura de todas as séries no site do Ipea
# Caso a conexão acabar, o código confere quais séreis ele já baixou e parte pras próximas
# Se cair a conexão, só rodar de novo

from os import listdir
from os.path import isfile, join
mypath = 'code_data'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

onfiles = []
for i in onlyfiles:
    onfiles.append(i[:-4])

time_in = [0]*len(test_lu)

# First Update
k = 0
for i in test_lu["CODE"]:
    now = datetime.now()
    if not i in onfiles:
        idpy.timeseries(i).to_csv("code_data\\" + i + ".csv", sep=",")
    now2 = datetime.now()
    seg = float(str(now2-now)[5:])
    time_in[k] = seg
    print('Seg: ' + str(seg) + '| Conc: ' + str(round((k+1)/len(test_lu)*100,2)))
    if k > 5:
        med_temp = (sum(time_in[k-4:k])/5)*(len(test_lu)-k)
        print('Estimativa: ' + str(round(med_temp/60,2)))
    k += 1
```

```python
# Parte demorada do código, captura as chaves de todas as séries e monta uma df mestre
# Caso a conexão acabar, o código confere quais séreis ele já baixou e parte pras próximas
# Se cair a conexão, só rodar de novo

time_in = [0]*len(test_lu)
k = 0

for i in test_lu["CODE"]:
    now = datetime.now()
    if k == 0:
        if not 'df_m' in globals():
            df_m = idpy.describe(i).transpose()
    else:
        if not i in list(df_m["Code"]):
            df_m = df_m.append(idpy.describe(i).transpose())
    now2 = datetime.now()
    seg = float(str(now2-now)[5:])
    time_in[k] = seg
    print('Seg: ' + str(seg) + '| Conc: ' + str(round((k+1)/len(test_lu)*100,2)))
    if k > 5:
        med_temp = (sum(time_in[k-4:k])/5)*(len(test_lu)-k)
        print('Estimativa: ' + str(round(med_temp/60,2)))
    k += 1
```
